```python
jr_list = ['Junior Financial Analyst', 'Junior HR Generalist', 'Junior Marketing Coordinator', 'Junior Marketing Specialist', 'Junior Operations Manager', 'Junior Project Manager', 'Junior Software Engineer']

jrmask = salaries['job_title'].isin(jr_list)
jr_data = salaries[jrmask]
jr_data
```

```python
jr_data_sorted = jr_data.pivot_table(index='job_title', columns='gender', values='salary')
jr_data_sorted = jr_data_sorted.reset_index()
jr_data_sorted
```

```python
jr_data_sorted_total = jr_data.pivot_table(index='job_title', values='salary')
jr_data_sorted_total = jr_data_sorted_total.reset_index()
jr_data_sorted_total
```

```python
sr_list = ['Senior Financial Analyst', 'Senior HR Generalist', 'Senior Marketing Coordinator', 'Senior Marketing Specialist', 'Senior Operations Manager', 'Senior Project Manager', 'Senior Software Engineer']

srmask = salaries['job_title'].isin(sr_list)
sr_data = salaries[srmask]
sr_data
```

```python
sr_data_sorted = sr_data.pivot_table(index='job_title', columns='gender', values='salary')
sr_data_sorted = sr_data_sorted.reset_index()
sr_data_sorted
```

```python
sr_data_sorted_total = sr_data.pivot_table(index='job_title', values='salary')
sr_data_sorted_total = sr_data_sorted_total.reset_index()
sr_data_sorted_total
```

```python
titles = ['Financial Analyst', 'HR Generalist', 'Marketing Coordinator', 'Marketing Specialist', 'Operations Manager', 'Project Manager', 'Software Engineer']

jrsr_data = pd.DataFrame({"Junior Title": jr_data_sorted["job_title"], "Senior Title": sr_data_sorted["job_title"]})

# Calculate the increase for "Male" by comparing "jr_data_sorted" with itself
jrsr_data["Male_Increase"] = sr_data_sorted["Male"] - jr_data_sorted["Male"]

# Calculate the increase for "Female" by comparing "jr_data_sorted" with "sr_data_sorted"
jrsr_data["Female_Increase"] = sr_data_sorted["Female"] - jr_data_sorted["Female"]

# Calculate the overall increase as the average of "Male Increase" and "Female Increase"
jrsr_data["Overall_Increase"] = sr_data_sorted_total["salary"] - jr_data_sorted_total["salary"]

# Calculate the percentage increase for "Male Increase"
jrsr_data["Male_Percentage"] = (jrsr_data["Male_Increase"] / jr_data_sorted["Male"])

# Calculate the percentage increase for "Female Increase"
jrsr_data["Female_Percentage"] = (jrsr_data["Female_Increase"] / jr_data_sorted["Female"])

# Calculate the percentage increase for "Overall Increase"
jrsr_data["Overall_Percentage"] = (jrsr_data["Overall_Increase"] / jr_data_sorted_total["salary"]) 

jrsr_data['Male_Increase'] = jrsr_data['Male_Increase'].fillna(0).astype(int)
jrsr_data['Female_Increase'] = jrsr_data['Female_Increase'].fillna(0).astype(int)
jrsr_data['Overall_Increase'] = jrsr_data['Overall_Increase'].fillna(0).astype(int)

jrsr_data['Titles'] = titles
jrsr_data
```

```python
json_data8 = jrsr_data.to_json(orient='records')
with open('data\data8_jr_sr_data.js', 'w') as file:
    file.write(f'let data8 = {json_data8};')        
```
